## PCA

```python
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

%matplotlib inline
import pandas as pd
import mglearn
```

```python
cancer = datasets.load_breast_cancer()

df = pd.DataFrame(cancer.data)
df.columns= cancer.feature_names
df["cancer"] = cancer.target # 0 tumor benigno y 1 tumor maligno
df.head()
```

## Sin escalar

```python
pca =PCA(n_components=3) #Cantidad de dimensiones a reducir
pca.fit(cancer.data)

transformada = pca.transform(cancer.data)

mglearn.discrete_scatter(transformada[:,0], transformada[:,1], cancer.target)
plt.legend(cancer.target_names, loc="best")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.show()
```

### Con data Escala

```python
escala = MinMaxScaler()
escalada = escala.fit_transform(cancer.data)

pca.fit(escalada)
transformada = pca.transform(escalada)

mglearn.discrete_scatter(transformada[:,0], transformada[:,1], cancer.target)
plt.legend(cancer.target_names, loc="best")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.show()
```

Se puede observar que con la data escala se puede visualizar mejor los tipos de tumores
